```python
import random
import math
import matplotlib.pyplot as plt

def bitString(l):
    bitList = []
    for i in range(0, l):
        temp = int(random.randint(0, 1))
        bitList.append(temp)
    return bitList

def invertBitString(bitString, mutationRate):
    for i in range(len(bitString)):
        rnd = random.randint(1,100)
        if rnd <= mutationRate:
            bitString[i] ^= 1
    return bitString

def compareBitStrings(x, xm, modificationBool):
    if modificationBool == False:
        if sum(x) < sum(xm):
            return xm
        else:
            return x
    else:
        return xm


def algorithm(l, mutationRate, iterations, modificationBool): 
    x = bitString(l)
    fitness = []
    optimumCount = 0
    for i in range(iterations):
        #print("x =  ", x, sum(x))
        xm = x.copy()
        xm = invertBitString(xm, mutationRate)
        #print("xm = ",xm, sum(xm))
        x = compareBitStrings(x, xm, modificationBool)
        #print("win =",x, sum(x))
        #print("")
        fitness.append(sum(x))
        
        if(sum(x) == len(x)):
            optimumCount += 1
            
    print("The optimum is found ", optimumCount, " times")
    return fitness
        
    
    
    
    
    
    
    
    
    
    
    
    
    

print("a")

plt.figure().set_figwidth(100)
plt.figure().set_figheight(5)

plt.plot(algorithm(l = 100, mutationRate = 1, iterations = 1500, modificationBool = False))
plt.ylabel("Fitness")
plt.xlabel("Iteration #")
plt.show()


print("b")

plt.figure().set_figwidth(100)
plt.figure().set_figheight(5)

for i in range(10):
    plt.plot(algorithm(l = 100, mutationRate = 1, iterations = 1500, modificationBool = False))
plt.ylabel("Fitness")
plt.xlabel("Iteration #")
plt.show()

print("c")

plt.figure().set_figwidth(100)
plt.figure().set_figheight(5)

for i in range(10):
    plt.plot(algorithm(l = 100, mutationRate = 1, iterations = 1500, modificationBool = True))
plt.ylabel("Fitness")
plt.xlabel("Iteration #")
plt.show()


```
