```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def PSO(func, n, a1, a2, r1, r2, w, X = 20, V = 10, maxIter = 10):
    
    #No distinction between local/global best, since both are essentially the same when n=1.
    best = X
    bestArray = []
    curPos = []
    curPos.append(X)
    gens = np.array(range(maxIter+1))
    
    bestArray.append(X)
    for i in (range(maxIter)):
        V = w*V + a1*r1*(best-X) + a2*r2*(best-X)
        X = X + V
        curPos.append(X)
        #update best
        if(func(X) < func(best)):
            best = X
        bestArray.append(best)
        
    return bestArray, gens, curPos
```

```python
def square(x):
    return x**2

x, gens, curPos = PSO(square,
                n = 1,
                w = 0.5,
                a1 = 1.5,
                a2 = 1.5,
                r1 = 0.5,
                r2 = 0.5,
                maxIter = 10,
)
locs, labels = plt.xticks()
plt.title("Best fitness per generation (w = .5)")
plt.plot(gens, curPos)
plt.ylabel("Current position")
plt.xlabel("Generation #")
plt.xticks(np.arange(0, 11, step=1))
plt.show()
print("Figure 1")

x, gens, curPos = PSO(square,
                n = 1,
                w = 0.7,
                a1 = 1.5,
                a2 = 1.5,
                r1 = 1,
                r2 = 1,
                maxIter = 10,
)

plt.title("Best fitness per generation (w = .7)")
plt.plot(gens, curPos)
plt.ylabel("Current position")
plt.xlabel("Generation #")
plt.xticks(np.arange(0, 11, step=1))
plt.show()
print("Figure 2")
```

**Discuss how would a particle ”swarm” consisting of a single member perform in such a minimization task when ω < 1, assuming the particle starts with the velocity pointing away from the optimum.**

When the swarm consists of just one particle, and this particle faces away from the optimum, the particle has to find the optimum on its own.  As can be observed in the figures above, $\omega$ < 1 in both cases. Thus, the velocity decreases over time. A single particle reaches the optimum when $\omega$ = 0.7 (Figure 2). However, a single particle does not reach the optimum when $\omega$ = 0.5 (Figure 1). Thus, depending on $\omega$, the particle reaches the optimum. 

When the velocity of a particle reaches zero, it is still able to move to a previous global best. This implies that the particle can still revert its position to the previous best. As such, the particle will only stop moving when the particle is facing the optimum, and velocity reaches 0 due to $\omega$ < 1.
