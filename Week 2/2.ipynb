```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def PSO(func, n, a1, a2, r1, r2, w, X = 20, V = 10, maxIter = 10):
    
    #No distinction between local/global best, since both are essentially the same when n=1.
    best = X
    bestArray = np.zeros(10)
    gens = np.array(range(maxIter))+1
    for i in (range(maxIter)):
        V = w*V + a1*r1*(best-X) + a2*r2*(best-X)
        X = X + V
        #print("velocity ",V)
        #print("position ",X)
        #print("Best: ",best) 
        
        #update best
        if(func(X) < func(best)):
            best = X
        bestArray[i] = best
        
    return bestArray, gens
```

```python
def square(x):
    return x**2

x, gens = PSO(square,
                n = 1,
                w = 0.5,
                a1 = 1.5,
                a2 = 1.5,
                r1 = 0.5,
                r2 = 0.5,
                maxIter = 10,
)

plt.title("Best fitness per generation (w = .5)")
plt.plot(gens, x)
plt.ylabel("Best fitness")
plt.xlabel("Generation #")
plt.show()

x, gens = PSO(square,
                n = 1,
                w = 0.7,
                a1 = 1.5,
                a2 = 1.5,
                r1 = 1,
                r2 = 1,
                maxIter = 10,
)

plt.title("Best fitness per generation (w = .7)")
plt.plot(gens, x)
plt.ylabel("Best fitness")
plt.xlabel("Generation #")
plt.show()
```
